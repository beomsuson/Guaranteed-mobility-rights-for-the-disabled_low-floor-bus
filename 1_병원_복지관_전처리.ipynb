```python
import pandas as pd
import sys

print(sys.version)
```

```python
##병원 raw데이터

```

```python
hospital=pd.read_excel('병원_raw_data.xlsx')
hospital
```

```python
columns=['기관명','주소','병원분류명','병원경도','병원위도'] #필요 컬럼 선별
hospital=hospital[columns]
hospital.head()
```

```python
## 병원 분류 : 종합 분류와 보건소만 취급
```

```python
filter=hospital['병원분류명']=='종합병원'
hospital_complex=hospital[filter]
filter=hospital['병원분류명']=='보건소'
hospital_facility=hospital[filter]
hospital=pd.concat([hospital_complex,hospital_facility]).reset_index().drop('index',axis=1)
hospital.head()
```

```python
hospital.to_csv('hospital.csv',index=False)
```
