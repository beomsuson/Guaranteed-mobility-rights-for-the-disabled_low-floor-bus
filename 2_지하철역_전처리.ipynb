```python
##카카오 api 활용 주소를 통해 좌표 출력
```

```python
import json
import requests

def to_point(address):
    api_key='c4b8b4ca703530438f9ee053e7777d81'
    url = 'https://dapi.kakao.com/v2/local/search/address.json?query={}'.format(address)
    headers = {"Authorization": "KakaoAK " + api_key}
    result = json.loads(str(requests.get(url, headers=headers).text))
    match_first = (float(result['documents'][0]['y']),float(result['documents'][0]['x'])) #[0]['address_name']
    return match_first

def to_point_json(address):
    api_key='c4b8b4ca703530438f9ee053e7777d81'
    url = 'https://dapi.kakao.com/v2/local/search/address.json?query={}'.format(address)
    headers = {"Authorization": "KakaoAK " + api_key}
    result = json.loads(str(requests.get(url, headers=headers).text))
    #match_first = (float(result['documents'][0]['y']),float(result['documents'][0]['x'])) #[0]['address_name']
    return result
```

```python
point_lst=[] ##좌표
goo_lst=[] ## 행정구
for j in range(len(subway)):
        df=subway.iloc[j]
        sub_name=df['역명']
        index=df['도로명주소'].find('(')
        address=df['도로명주소'][:index]
        goo=address.split()[1]
        
        goo_lst.append(goo)
        try:
            sub=to_point(address)
            point_lst.append(sub)
        except:
            sub=to_point_json(address)
            point_lst.append(sub)

            
      
```

```python
subway['좌표']=point_lst
subway['행정구']=goo_lst
```
